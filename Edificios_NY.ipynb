- CONVERTIR LOS PIES CUADRADOS A METROS CUADRADOS

```python
df2['PIES_CUADRADOS_TERRENO'] = pd.to_numeric(df2['PIES_CUADRADOS_TERRENO'], errors='coerce')
df2['PIES_CUADRADOS_BRUTOS'] = pd.to_numeric(df2['PIES_CUADRADOS_BRUTOS'], errors='coerce')


df2['METROS_CUADRADOS_TERRENO'] = df2['PIES_CUADRADOS_TERRENO'] * 0.092903
df2['METROS_CUADRADOS_BRUTOS'] = df2['PIES_CUADRADOS_BRUTOS'] * 0.092903


index = df2.columns.get_loc('PIES_CUADRADOS_TERRENO') + 1
df2.insert(index, 'METROS_CUADRADOS_TERRENO', df2.pop('METROS_CUADRADOS_TERRENO'))

index = df2.columns.get_loc('PIES_CUADRADOS_BRUTOS') + 1
df2.insert(index, 'METROS_CUADRADOS_BRUTOS', df2.pop('METROS_CUADRADOS_BRUTOS'))

df2
```

```python
df3 = df2.drop(columns={"PIES_CUADRADOS_TERRENO", "PIES_CUADRADOS_BRUTOS"})
df3
```

```python
df3["SECTOR"] = df3["SECTOR"].replace({1: "MANHATTAN", 2: "BRONX", 3: "BROOKLYN", 4: "QUEENS", 5: "STATEN ISLAND"})
df3
```

```python
# Calcula los cuantiles
Q1 = df3['AÑO_CONSTRUCCION'].quantile(0.25)
Q3 = df3['AÑO_CONSTRUCCION'].quantile(0.75)

# Calcula el rango intercuartílico (IQR)
IQR = Q3 - Q1

# Calcula los límites inferior y superior para identificar valores atípicos
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtra los valores atípicos
outliers = df3[(df3['AÑO_CONSTRUCCION'] < lower_limit) | (df3['AÑO_CONSTRUCCION'] > upper_limit)]

# Muestra los valores atípicos
print("Valores atípicos en AÑO_CONSTRUCCION:")
print(outliers)

```

# CONCLUSIONES

Las conclusiones extraídas del análisis del conjunto de datos inmobiliarios proporcionan una visión detallada del mercado de propiedades, destacando las siguientes observaciones clave:

Alta Actividad Inmobiliaria: El mercado se caracteriza por una alta actividad, con 15,446 propiedades registradas, indicando un sector vibrante con una amplia oferta de propiedades.

Preferencia Histórica y Conservación: Un pico notable en el año de construcción más común, 1920, sugiere una era de desarrollo significativo o un interés en la conservación de propiedades de este periodo, reflejando la riqueza histórica del área.

Dominio de Brooklyn y Manhattan: Brooklyn sobresale por tener la mayor cantidad de propiedades, mientras que Manhattan se distingue por el promedio más alto de áreas totales y por albergar la propiedad con el precio de venta más alto, subrayando la diversidad y especialización geográfica del mercado.

Variedad de Propiedades: El rango de propiedades, desde aquellas con amplias áreas hasta aquellas con precios de venta excepcionalmente altos o bajos, muestra un mercado diverso que puede acomodar a diferentes grupos de compradores e inversores.

Crecimiento y Movilidad: La venta de propiedades casi duplicada de un año a otro sugiere un crecimiento significativo o una mayor movilidad en el mercado, lo que podría indicar un momento oportuno para la inversión o la venta.

Mercado de Lujo: El alto precio de venta promedio de las propiedades refleja la presencia de un segmento de mercado de lujo, con propiedades de alto valor que atraen a compradores e inversores con presupuestos sustanciales.

Oportunidades de Inversión en Bedford Stuyvesant: Este barrio no solo tiene la mayor cantidad de propiedades registradas sino también la mayor cantidad de propiedades en venta, marcándolo como un punto caliente para la actividad inmobiliaria y potencialmente una buena oportunidad de inversión.

Histórico y Moderno: La presencia de propiedades que datan desde 1800 hasta 2017 ilustra la mezcla única de lo histórico y lo contemporáneo en el mercado, ofreciendo opciones para aquellos interesados en propiedades con carácter histórico o moderno.

Estas conclusiones enfatizan un mercado inmobiliario dinámico, con oportunidades para diversos intereses y necesidades, desde inversionistas hasta compradores en busca de residencias con características específicas. La diversidad geográfica, histórica, y de precios subraya la riqueza y complejidad del mercado, presentando un paisaje interesante para la exploración y la inversión futura.
